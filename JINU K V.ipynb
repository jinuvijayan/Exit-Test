# Outliers

```python
# Define a function to detect outliers using Z-score
def find_outliers_zscore(data, threshold=3):
    z_scores = np.abs((data - data.mean()) / data.std())
    return z_scores > threshold
```

```python
outliers_all_columns = data.drop(columns=['player_id', 'name','selection','gender','country','height','weight','strong_foot','dedication_level','coaching']).apply(find_outliers_zscore)
```

```python
# Show the rows where any column has an outlier
outliers_rows = outliers_all_columns.any(axis=1)
outliers_detected = data[outliers_rows]
```

```python
# Get the index of potential outliers
potential_outliers_index = outliers_detected.index
```

```python
potential_outliers_index
```

```python
# Remove the potential outliers from the original DataFrame
df_cleaned = data.drop(index=potential_outliers_index)
```

```python
# After removing outliers and storing the cleaned DataFrame in 'df_cleaned'
# Reset the index of the DataFrame
df_cleaned.reset_index(drop=True, inplace=True)
```

## Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df_cleaned['gender'] = le.fit_transform(df_cleaned['gender'])
df_cleaned['height'] = le.fit_transform(df_cleaned['height'])
df_cleaned['weight'] = le.fit_transform(df_cleaned['weight'])
df_cleaned['strong_foot'] = le.fit_transform(df_cleaned['strong_foot'])
df_cleaned['dedication_level'] = le.fit_transform(df_cleaned['dedication_level'])
df_cleaned['coaching'] = le.fit_transform(df_cleaned['coaching'])
```

```python
df_cleaned
```

```python
df_cleaned.drop(["player_id","name","country"],axis=1,inplace=True)
```

```python
df_cleaned
```

```python
## To spilt into two variables
x=df_cleaned.drop(['selection'],axis=1)
y=df_cleaned['selection']
```

## Scaling

```python
## To do minmax scaler
from sklearn.preprocessing import MinMaxScaler
min_max=MinMaxScaler(feature_range=(0,1))
x=min_max.fit_transform(x)
```

```python
x
```

```python
## To change array to Dataframe
x=pd.DataFrame(x)
```

## train_test_split

## Logistic Regression

```python
## To split data in training and test sets and to apply logistic regression
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
from sklearn.linear_model import LogisticRegression
Ir=LogisticRegression()
model=Ir.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
# Evaluate the model
from sklearn.metrics import accuracy_score, classification_report
print("Logistic Regression:")
print("Accuracy:", accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

## Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
# Train the Decision Tree model
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(x_train, y_train)
```

```python
# Make predictions on the test set
dt_y_pred = dt_model.predict(x_test)
```

```python
# Evaluate the model
print("Decision Tree:")
print("Accuracy:", accuracy_score(y_test, dt_y_pred))
print(classification_report(y_test, dt_y_pred))
```

## Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
# Train the Random Forest model
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(x_train, y_train)
```

```python
# Make predictions on the test set
rf_y_pred = rf_model.predict(x_test)
```

```python
# Evaluate the model
print("Random Forest:")
print("Accuracy:", accuracy_score(y_test, rf_y_pred))
print(classification_report(y_test, rf_y_pred))
```

## KNN Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
# create an empty list.
metric_k=[]
neighbors=np.arange(3,15)
```

```python
# Train the KNN model
for k in neighbors:
  knn_model=KNeighborsClassifier(n_neighbors=k,metric='euclidean')
  knn_model.fit(x_train,y_train)
  # Make predictions on the test set
  knn_y_pred=knn_model.predict(x_test)
  acc=accuracy_score(y_test,knn_y_pred)
  metric_k.append(acc)
```

```python
# Lets plot a graph to find highest accuracy score of given k
plt.plot(neighbors,metric_k,'o-')
plt.xlabel('k')
plt.ylabel('Accuracy')
plt.grid()
```

```python
# Lets create knn model for 4 neighbors
knn_model=KNeighborsClassifier(n_neighbors=4,metric='euclidean')
knn_model.fit(x_train,y_train)
```

```python
# Make predictions on the test set
knn_y_pred=knn_model.predict(x_test)
```

```python
# Evaluate the model
print("KNN Model:")
print("Accuracy:", accuracy_score(y_test, knn_y_pred))
print(classification_report(y_test, knn_y_pred))
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
param_dist = {
    'n_neighbors': range(3, 15),
    'weights': ['uniform', 'distance'],
    'p': [1, 2]
}
```

```python
# perform randomized search cv
random_search = RandomizedSearchCV(
    knn_model,
    param_distributions=param_dist,
    n_iter=10,  # Number of parameter settings to sample
    cv=2,       # Cross-validation folds
    random_state=42
)

random_search.fit(x_train, y_train)
```

```python
# Print best parameters
print("Best parameters:", random_search.best_params_)
```

```python
# Make predictions
best_knn = random_search.best_estimator_

predictions = best_knn.predict(x_test)
```

```python
# Evaluate the model
print("Randomsearch Model:")
print("Accuracy:", accuracy_score(y_test, predictions))
print(classification_report(y_test, predictions))
```
